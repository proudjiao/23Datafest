```python
# questionposts = pd.read_csv('questionposts.csv', low_memory=False)
questions['ClosedOnUtc'] = pd.to_datetime(questions['ClosedOnUtc'])
questions['AskedOnUtc'] = pd.to_datetime(questions['AskedOnUtc'])
questions['TakenOnUtc'] = pd.to_datetime(questions['TakenOnUtc'])
```

```python
import matplotlib.pyplot as plt

# Group by year and category and compute proportions
questions_by_year = questions.groupby([questions['ClosedOnUtc'].dt.year, 'Category'])['Category'].size().unstack()
questions_by_year_prop = questions_by_year.apply(lambda x: x/x.sum(), axis=1)

# Plot the stacked bar chart
ax = questions_by_year_prop.plot(kind='bar', stacked=True)

# Set the chart title and axis labels
ax.set_title('Questions by Year and Category')
ax.set_xlabel('Year')
ax.set_ylabel('Proportion')

# Move the legend to the right
ax.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))

# Show the chart
plt.show()
```
